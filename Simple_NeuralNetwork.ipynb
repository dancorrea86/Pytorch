```python
import torch
import torch.nn as nn
import torch.nn.functional as F
```

```python
# Create a Model Class that inherits nn.Module

class Model(nn.Module):
  # Input layer (4 tipos de flores) "in_features" -->
  # 1 Camada escondida (nmero de neurnios) "h1" -->
  # 2 Camada escondida (nmero de neurnio) "h2" -->
  # resultado (3 classe de Iris Flores) out_features
  def __init__(self, in_features=4, h1=8, h2=9, out_features=3):
    super().__init__()
    self.fc1 = nn.Linear(in_features, h1) # Ligao do input com a 1 camada
    self.fc2 = nn.Linear(h1, h2) # Ligao 1 camada com 2 camada
    self.out = nn.Linear(h2, out_features) # Ligao 2 camada com output



```
