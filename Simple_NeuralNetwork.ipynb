```python
import torch
import torch.nn as nn
import torch.nn.functional as F
```

```python
# Create a Model Class that inherits nn.Module

class Model(nn.Module):
  # Input layer (4 tipos de flores) "in_features" -->
  # 1º Camada escondida (número de neurônios) "h1" -->
  # 2º Camada escondida (número de neurônio) "h2" -->
  # resultado (3 classe de Iris Flores) out_features
  def __init__(self, in_features=4, h1=8, h2=9, out_features=3):
    super().__init__()
    self.fc1 = nn.Linear(in_features, h1) # Ligação do input com a 1º camada
    self.fc2 = nn.Linear(h1, h2) # Ligação 1º camada com 2º camada
    self.out = nn.Linear(h2, out_features) # Ligação 2º camada com output



```
